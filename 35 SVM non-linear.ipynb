```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("C:/Users/sumit/OneDrive/Documents/polynomial_classification_data.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(data=dataset,x="data1",y="data2",hue="output")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["output"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 , random_state =38)
```

```python
from sklearn.svm import SVC
```

```python
sv = SVC(kernel = "sigmoid")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100 , sv.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() ,clf=sv)
plt.show()
```
