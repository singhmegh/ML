```python
new_data = dataset[['Item 1', 'Item 2', 'Item 3 ', 'Item 4', 'Item 5', 'Item 6', 'Item 7',
       'Item 8', 'Item  9', 'Item 10 ', 'Item 11']].copy()
```

# make a list of lists in market

```python
market = []
for i in range (0,new_data.shape[0]):           #Loops through each row 
    cus =[]
    for j in new_data.columns:                  #Loops through each columns
        if type(new_data[j][i])==str:
            cus.append(new_data[j][i])
    market.append(cus)
```

# make single list by market

```python
l = []
for i in market:                  #Har customer ki shopping list (i) lega.
    for j in i:                   #Us shopping list ke har item (j) ko lega
        l.append(j)
```

```python
import collections
```

```python
p = collections.Counter(l)
```

```python
d = {"Item Name":p.keys(),"Values":p.values()}
```

```python
pd.DataFrame(d).sort_values(by=["Values"],ascending = False)
```
