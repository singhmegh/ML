```python
import pandas as pd
```

```python
dataset = pd.read_csv("C:/Users/sumit/Downloads/placement.csv")
```

```python
dataset.head(3)
```

```python
x = dataset.iloc[:,:-1]
y= dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)*100 , lr.score(x_test,y_test)*100
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
dt.score(x_train,y_train)*100 , dt.score(x_test,y_test)*100
```

```python
sv = SVR()
sv.fit(x_train,y_train)
sv.score(x_train,y_train)*100 , sv.score(x_test,y_test)*100
```

# Voting Regressor model

```python
from sklearn.ensemble import VotingRegressor
```

```python
li = [("lr1",LinearRegression()),("dt1",DecisionTreeRegressor()),("sv1",SVR())]
```

```python
vr = VotingRegressor(li)
vr.fit(x_train,y_train)
```

```python
vr.score(x_train,y_train)*100 , vr.score(x_test,y_test)*100
```

```python
df = {"lr":lr.predict(x_test),"dt":dt.predict(x_test),"svm":sv.predict(x_test),"vr":vr.predict(x_test)}
```
