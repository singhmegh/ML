```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Simple Thersholding 

```python
img = cv2.imread("C:/Users/sumit/OneDrive/Desktop/cvimg/coin.jpg")
img = cv2.resize(img,(300,300))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_,th = cv2.threshold(gray,190,255,cv2.THRESH_BINARY)


cv2.imshow("Th_coin",th)
cv2.imshow("coin",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Otsu's Binarization Thesholding

```python
img = cv2.imread("C:/Users/sumit/OneDrive/Desktop/cvimg/noise leaf.jpeg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([img],[0],None,[256],[0,255])

plt.plot(hist)
plt.title("Grayscale Histogram")
plt.xlabel("Intensity Value")
plt.ylabel("Count")
plt.show()

_,th = cv2.threshold(img,120,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

cv2.imshow("leaf",img)
cv2.imshow("Otsu_leaf",th)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
