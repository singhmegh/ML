# Otsu's Binarization Thesholding

```python
img = cv2.imread("C:/Users/sumit/OneDrive/Desktop/cvimg/noise leaf.jpeg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([img],[0],None,[256],[0,255])

plt.plot(hist)
plt.title("Grayscale Histogram")
plt.xlabel("Intensity Value")
plt.ylabel("Count")
plt.show()

_,th = cv2.threshold(img,120,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

cv2.imshow("leaf",img)
cv2.imshow("Otsu_leaf",th)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# CLAHE (Contrast Limited Adaptive Histogram Equalization)

```python
img = cv2.imread("C:/Users/sumit/OneDrive/Desktop/cvimg/Clahe.jpeg")
img = cv2.resize(img,(500,500))

# use of Equalizer histogram
gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([gry],[0],None,[256],[0,255])
eq = cv2.equalizeHist(gry)

hist1 = cv2.calcHist([eq],[0],None,[256],[0,255])

plt.plot(hist1,c="red")
plt.plot(hist)
plt.show()

cv2.imshow("eq_img",gry)
cv2.imshow("CLAHE",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("C:/Users/sumit/OneDrive/Desktop/cvimg/Clahe.jpeg")
img = cv2.resize(img,(500,500))

# use of Equalizer histogram
gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([gry],[0],None,[256],[0,256])
eq = cv2.equalizeHist(gry)

hist1 = cv2.calcHist([eq],[0],None,[256],[0,256])

# use CLAHE

cl = cv2.createCLAHE(clipLimit = 1, tileGridSize = (8,8))
f =  cl.apply(gry)


plt.plot(hist1,c="red")
plt.plot(hist)
plt.plot(f,c = "yellow")
plt.show()

cv2.imshow("clahe_img1",f)
cv2.imshow("eq_img",gry)
cv2.imshow("org",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
