```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("C:/Users/sumit/Downloads/polynomial_level_salary_1000_integer.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
plt.scatter(dataset["Level"],dataset["Salary"])
plt.ylabel("Salary")
plt.xlabel("Level")
plt.show()
```

```python
dataset.corr()
```

```python
x = dataset[["Level"]]
y = dataset["Salary"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree = 2)
pf.fit(x)
x = pf.transform(x) 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
#y=m1x1+m2x2^2+c

#y = 502.73284581*x1 + 99.995379398*x2^2 + 1748.6653405241668
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
prd = lr.predict(x)
```

```python
plt.scatter(dataset["Level"],dataset["Salary"])
plt.plot(dataset["Level"],prd,c="red")
plt.ylabel("Salary")
plt.xlabel("Level")
plt.legend(["org","prd"])
plt.show()
```
