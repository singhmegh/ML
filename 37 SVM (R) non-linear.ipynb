```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("C:/Users/sumit/Downloads/polynomial_level_salary_1000_integer.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(data=dataset,x="Level",y="Salary")
plt.show()
```

```python
x=dataset[["Level"]]
y=dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 ,random_state = 42)
```

```python
from sklearn.svm import SVR
```

```python
sv = SVR(kernel = "linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100 , sv.score(x_train,y_train)*100
```

```python
sns.scatterplot(data=dataset,x="Level",y="Salary")
plt.plot(dataset["Level"],sv.predict(x),color = "red")
plt.show()
```
