```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("C:/Users/sumit/OneDrive/Desktop/Social_Network_Ads.csv")
dataset.head(3)
```

```python
sns.scatterplot(data=dataset,x="Age",y="EstimatedSalary",hue="Purchased")
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Purchased"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)*100 , knn.score(x_train,y_train)*100
```

```python
for i in range(1,30):
    knn1= KNeighborsClassifier(n_neighbors = i)
    knn1.fit(x_train,y_train)
    print(i, knn1.score(x_train,y_train)*100, knn1.score(x_test,y_test)*100)
```
