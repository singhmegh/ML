```python
dataset = pd.read_csv("C:/Users/sumit/Downloads/diabetes.csv")
dataset.head(5)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Outcome"]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
fs = SequentialFeatureSelector(lr,k_features =6,forward=True )
fs.fit(x,y)
```
