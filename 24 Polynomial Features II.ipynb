```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("C:/Users/sumit/OneDrive/Documents/polynomial_classification_data.csv")
dataset.head(3)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(data=dataset,x="data1",y="data2",hue="output")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["output"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Assume 'x' is your original DataFrame before polynomial transformation
pf = PolynomialFeatures(degree=2, include_bias=False)

# Save column names BEFORE transform
original_columns = x.columns

# Fit and transform
x_poly = pf.fit_transform(x)

# Get the new column names
feature_names = pf.get_feature_names_out(input_features=original_columns)

# Create new DataFrame
x = pd.DataFrame(x_poly, columns=feature_names)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr= LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```
