```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("C:/Users/sumit/OneDrive/Desktop/Social_Network_Ads.csv")
dataset.drop(columns=["EstimatedSalary"],inplace=True)
```

```python
dataset.head(3)
```

```python
dataset["Age"].astype(int)
```

```python
sns.scatterplot(data=dataset,x="Age",y="Purchased")
plt.show()
```

```python
x=dataset[["Age"]]
y=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[40]])
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(data=dataset,x="Age",y="Purchased")
sns.lineplot(data=dataset,x="Age",y=lr.predict(x),color ="red")
plt.show()
```
