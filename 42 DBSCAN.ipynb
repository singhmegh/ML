```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples = 250,noise = 0.05)
```

```python
#noise=0.05	Adds a small amount of random noise to make the data slightly messy
```

```python
df = {"data1":x[:,0],"data2":x[:,1]}
```

```python
dataset = pd.DataFrame(df)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(data=dataset,x="data1",y="data2")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps = 0.2,min_samples = 5)
dataset["Predict"] = db.fit_predict(dataset)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(data=dataset,x="data1",y="data2",hue ="Predict")
plt.show()
```

```python
sns.scatterplot(data=dataset,x="data1",y="data2",hue = y)
plt.show()
```
